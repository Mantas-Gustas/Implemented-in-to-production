```python
from module_files.funcs_vars import *
import pyautogui as py
import pandas as pd
import time
import numpy as np
import keyboard 
import sys, os


data = pd.read_excel(r'B:\Tariffing\Tariff Build Master Spreadsheet LATEST VERSION 01-10-2021 - Auto Build Inc.xls', sheet_name='New ABS', dtype=str)

data.columns = data.iloc[0]  # copies row 0 to header
data = data.drop(0)  # drops/removes row 0 as its already copied to headers.

needed_data = data[data['Auto Build YES/DONE'] == 'YES']
needed_data = needed_data.dropna(axis = 'columns')
needed_data.reset_index()
```

```python
# -============= VARIABLE HOLDING VALUES & DATA REQUEST FROM EXCEL ============-


# -========= CHANGE DATE ==========-

date_from = '01022022000000'
gsm_network_code = 'O2'

# -========== VARIABLES ===========-

#  Fixed Charge Variable holding Values extracted from Excel
fc_code = pd.DataFrame(needed_data['FC Code'])
fc_code = fc_code.iloc[0]['FC Code']

fc_name = pd.DataFrame(needed_data['FC Name'])
fc_name = fc_name.iloc[0]['FC Name']

fc_cost = pd.DataFrame(needed_data['FC Price'])
fc_cost = fc_cost.iloc[0]['FC Price']

fc_profit = pd.DataFrame(needed_data['LR Profit'])
fc_profit = fc_profit.iloc[0]['LR Profit']

fc_nom_code = pd.DataFrame(needed_data['FC Nominal'])
fc_nom_code = fc_nom_code.iloc[0]['FC Nominal']

#  Template Variable holding Values extracted from Excel
temp_code = pd.DataFrame(needed_data['Template Code'])
temp_code = temp_code.iloc[0]['Template Code']

temp_desc = pd.DataFrame(needed_data['Template Name'])
temp_desc = temp_desc.iloc[0]['Template Name']

temp_air = pd.DataFrame(needed_data['Airlimit'])
temp_air = temp_air.iloc[0]['Airlimit']

uc_code = pd.DataFrame(needed_data['UC Code'])
uc_code = uc_code.iloc[0]['UC Code']

bund_code = pd.DataFrame(needed_data['Data Bundle Code'])
bund_code = bund_code.iloc[0]['Data Bundle Code']

#  O2 ABS Other Vases
cert_vas = pd.DataFrame(needed_data['Self Cert VAS'])
cert_vas = cert_vas.iloc[0]['Self Cert VAS']


# -========== ALL FUNCTIONS ADDED TOGETHER THAT EXECUTES THE BUILD ===========- 

# -=============== GSM Fixed Charge Code - Auto Build Code ==============-

def mobile_fixCharge():
    time.sleep(3)
    py.click(gsm_png)
    time.sleep(1)
    for i in range(3):
        down()
    py.press('space')
    time.sleep(2)
    tab()
    for fc in [fc_code, fc_name, fc_code]:
        py.write(fc)
        tab()
    for gr in ['AF ', 'O2ABSLR ']:
        py.write(gr, interval=0.02)
        backspace()
        tab()
    py.write(fc_cost, interval=0.02)
    tab()
    for i in range(3):
        tab()
    A()   
    for i in range(6):
        P()
    py.write(fc_profit, interval=0.02)
    tab()
    tab()
    py.write(fc_nom_code, interval=0.02)
    tab()
    space()
    f34a4()


#  -============ GETTING TO AND Building GSM Template "Details" Window =============-

def template_details():
    time.sleep(3)
    py.click(gsm_png), time.sleep(1)
    for i in range(6): 
        down()
    space()
    time.sleep(2)
    #  GSM Template "Details" Window
    for temp in [temp_code, temp_desc, date_from]:
        tab()
        py.write(temp, interval=0.02)
    for i in range(2):
        tab()
    py.write(gsm_network_code, interval=0.02)
    tab()
    onoo()
    for nt in ['TCP ', 'O2 ']:
        py.write(nt, interval=0.05)
        backspace()
        tab()
    O()       
    py.write(temp_air, interval=0.02)
    tab()
    for i in range(2): 
        O()
        tab()
    f3()

#  -=============== GSM "VAS" Window ==================- 

def template_vases():
    ctab()
    dett_butt()
    add_o2Vases()
    f4()   
    
# -============== GSM "Fix Charge" Window ==============-
    
def template_lineRental():
    for ct in range(2):
        ctab()
    cf3()
    tab()
    py.write('Line Rental', interval=0.02)
    for i in range(2):
        tab()
    py.write('IA ', interval=0.02)
    backspace()
    tab()
    py.write('O2PV001 ', interval=0.02)
    backspace()
    for i in range(5):
        tab()
    py.write(fc_code, interval=0.02)
    tab()
    cf3()
    f3()
    af4()


# -=============== GSM "USG Charge" Window ================-

def template_callCharges():
    py.click(f3_new_png), time.sleep(0.5)
    tab()
    py.write('Call Charges', interval=0.02)
    tab()
    py.write('UC ', interval=0.02)
    backspace()
    for i in range(2):
        tab()
    py.write('O2PV001 ', interval=0.02)
    backspace()
    for i in range(5):
        tab()
    py.write(uc_code, interval=0.02)
    tab()
    cf3()
    ctab()
    for png in [details_but_png, buttons_but_png, bundle_code_png]:
        py.click(png), time.sleep(0.02)
    py.write(bund_code, interval=0.02)
    tab()
    py.click(bundle_date_from_png)
    time.sleep(0.02)
    py.write(date_from, interval=0.02)
    tab()
    cf3()
    f3()
    af4()
    f4()

    
# # -=============== THE BELOW WILL BUILD FIX CHARGE CODE -=====================

if __name__ == "__main__":
    mobile_fixCharge()

# -=============== THE BELOW WILL BUILD TEMPLATE DETAILS WINDOW -=====================

    template_details()

# -=============== THE BELOW WILL BUILD TEMPLATE VAS WINDOW -=====================

    template_vases()

# -=============== THE BELOW WILL BUILD TEMPLATE LINE RENTAL WINDOW -=====================

    template_lineRental()

# -=============== THE BELOW WILL BUILD TEMPLATE CALL CHARGES WINDOW -=====================

    template_callCharges()
```
